```python
import xlrd
import numpy as np
```

```python
worksheet=xlrd.open_workbook('GuCheng.xlsx')#打开文件
```

```python
sheet_names=worksheet.sheet_names()#获取excel中所有工作表名
```

```python
sheet2=worksheet.sheet_by_name('Sheet2')#根据Sheet名获取数据
```

```python
rows=sheet2.row_values(3)#表示获取Sheet2中第4行数据
```

```python
cols3=sheet2.col_values(2)#表示获取Sheet2中第3列的数据
x = np.array(cols3)
column=x.shape[0]
column#列数
```

```python
List=[]
count=0
for i in range(2,column):
    Li=[]
    if cols3[i].find(',')==-1:#如果该列为空，则continue直接开始下一轮循环
        continue
    Li.append(float(cols3[i][cols3[i].find(',')+1:len(cols3[i])]))#纬度
    Li.append(float(cols3[i][0:cols3[i].find(',')]))#经度
    List.append(Li)
    count=count+1#记录List的长度
print(count)
List
```

```python
"""
Stamen Toner图
"""
import folium
m = folium.Map(location=[25.6929, 100.1740],
               tiles='Stamen Toner',
               zoom_start=14
)
for i in range(0,count):
    folium.features.Circle(
        radius=10,
        location=List[i],
        popup='The Waterfront',
        color='crimson',
        fill=True,
    ).add_to(m)
m
```
